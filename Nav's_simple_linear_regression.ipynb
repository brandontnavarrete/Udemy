# Simple Linear Regression

## Importing the libraries

```
# essential import
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
# read data and separate feats and dependent variable
dataset = pd.read_csv('Salary_Data.csv')
# selecting features
X = dataset.iloc[:, :-1].values
# selecting dependent variable
y = dataset.iloc[:, -1].values
```

## Splitting the dataset into the Training set and Test set

```
# method of splitting
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2, random_state = 1)
```

## Training the Simple Linear Regression model on the Training set

```
# library sklearn and module linear_model
from sklearn.linear_model import LinearRegression
# object instance
regressor = LinearRegression()
# fit the object on X_train and y_train
regressor.fit(X_train, y_train)
```

## Predicting the Test set results

```
# .predict on the "future data"
y_pred = regressor.predict(X_test)
```

```
print(y_pred)
```

## Visualising the Training set results

```
# scatter plot for x train and y train cordinates, dots are red
plt.scatter(X_train, y_train, color = 'green')
# regression line using plot method
plt.plot(X_train, regressor.predict(X_train), color = 'red')

plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Test set results

```
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'green')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
